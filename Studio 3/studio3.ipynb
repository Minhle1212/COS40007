Studio Activity 2 - Model training

```python
import pandas as pd

# Load the combined and shuffled data
all_data_df = pd.read_csv(data_path + '/all_data.csv')

original_rows = len(all_data_df)
sample_fraction = 0.30 
all_data_df = all_data_df.sample(frac=sample_fraction, random_state=42)
print(f"Original dataset had {original_rows} rows.")
print(f"Using a {sample_fraction*100}% sample with {len(all_data_df)} rows for faster training.")
# Separate features (X) and class (y)
# Assuming the last column is the class label
X = all_data_df.drop('class', axis=1)
y = all_data_df['class']

# Display the shapes of X and y to verify
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

Train an SVM model using splitting the train and test set to 70%/30% and measuring the model accuracy

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
clf_73 = svm.SVC()
clf_73.fit(X_train, y_train)
y_pred = clf_73.predict(X_test)
accuracy_score(y_test, y_pred)
```

Train an SVM model using 10-fold cross-validation and measure model accuracy (cross-validation score)

```python
from sklearn.model_selection import cross_val_score
from sklearn import svm
clf_cross = svm.SVC()
scores = cross_val_score(clf_cross, X, y, cv=10)
print(scores)
```

Studio Activity 7: Other Classifiers

Use the original data (all_data.csv) to
1) Train with SGDclassifier for both train-test split and cross-validation and obtain the accuracy value.
2) Train with RandomForest for both train-test split and cross-validation and obtain the accuracy value
3) Train with MLPclassifier for both train-test split and cross-validation and obtain the accuracy value

```python
# SGDclassifier
from sklearn.linear_model import SGDClassifier
clf_sgd = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)
clf_sgd.fit(X_train, y_train)
y_pred_sgd = clf_sgd.predict(X_test)
print(accuracy_score(y_test, y_pred_sgd))
```

```python
# RandomForest
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
clf_rf.fit(X_train, y_train)
y_pred_rf = clf_rf.predict(X_test)
print(accuracy_score(y_test, y_pred_rf))
```

```python
# MLPclassifier
from sklearn.neural_network import MLPClassifier
clf_mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
clf_mlp.fit(X_train, y_train)
y_pred_mlp = clf_mlp.predict(X_test)
print(accuracy_score(y_test, y_pred_mlp))
```
