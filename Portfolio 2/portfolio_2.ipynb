Step 2:Create composite columns

```python

import numpy as np

def calc_rmsq_for_cols(df: pd.DataFrame, cols: list[str]):
  return np.sqrt(np.mean(df[cols] ** 2, axis=1))

def calc_roll_for_col(df: pd.DataFrame, col : str):
    return 180 * np.arctan2(df[f'{col} y'], np.sqrt(df[f'{col} x']**2 + df[f'{col} z']**2)) / np.pi

def calc_pitch_for_col(df: pd.DataFrame, col : str):
    return 180 * np.arctan2(df[f'{col} x'], np.sqrt(df[f'{col} y']**2 + df[f'{col} z']**2)) / np.pi
```

Right hand feature


```python
step2_df = combined_df.copy()

step2_df['right_forearm_xy_rmsq'] = calc_rmsq_for_cols(step2_df, ['Right Forearm x', 'Right Forearm y'])
step2_df['right_forearm_yz_rmsq'] = calc_rmsq_for_cols(step2_df, ['Right Forearm y', 'Right Forearm z'])
step2_df['right_forearm_zx_rmsq'] = calc_rmsq_for_cols(step2_df, ['Right Forearm z', 'Right Forearm x'])
step2_df['right_forearm_xyz_rmsq'] = calc_rmsq_for_cols(step2_df, ['Right Forearm x','Right Forearm y', 'Right Forearm z'])
step2_df['right_forearm_roll'] = calc_roll_for_col(step2_df, 'Right Forearm')
step2_df['right_forearm_pitch'] = calc_pitch_for_col(step2_df, 'Right Forearm')
```

Left hand feature

```python
step2_df['left_forearm_xy_rmsq'] = calc_rmsq_for_cols(step2_df, ['Left Forearm x', 'Left Forearm y'])
step2_df['left_forearm_yz_rmsq'] = calc_rmsq_for_cols(step2_df, ['Left Forearm y', 'Left Forearm z'])
step2_df['left_forearm_zx_rmsq'] = calc_rmsq_for_cols(step2_df, ['Left Forearm z', 'Left Forearm x'])
step2_df['left_forearm_xyz_rmsq'] = calc_rmsq_for_cols(step2_df, ['Left Forearm x','Left Forearm y', 'Right Forearm z'])
step2_df['left_forearm_roll'] = calc_roll_for_col(step2_df, 'Left Forearm')
step2_df['left_forearm_pitch'] = calc_pitch_for_col(step2_df, 'Left Forearm')
```

Data pre-processing and Feature computation

```python
composited_raw_df = pd.read_csv('D:/COS40007/Portfolio 2/composited_data.csv')
composited_df = composited_raw_df.copy()
composited_df.info()
```

```python
from scipy.signal import find_peaks
from numpy import trapz


composited_df['minute'] = composited_df.index // 60

def calculate_auc(x):
    return trapz(x)

def count_peaks(x):
    peaks, _ = find_peaks(x)
    return len(peaks)

feature_cols = [col for col in composited_df.columns if col not in ['Frame', 'class', 'minute']]

df_features = composited_df.groupby('minute')[feature_cols].agg(
    ['mean', 'std', 'min', 'max', calculate_auc, count_peaks]
)

df_features.columns = ['_'.join(col).strip() for col in df_features.columns.values]


class_labels = composited_df.groupby('minute')['class'].first()
df_features['class'] = class_labels


print(f"Step 3 complete. The final feature set has {df_features.shape[1]-1} features and {df_features.shape[0]} samples.")

df_features.head()
```

```python
df_features.to_csv('D:/COS40007/Portfolio 2/final_features.csv', index= False)
print("Shape: ", df_features.shape)
print("Info: ", df_features.info())
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

Load csv

```python
all_data_df = pd.read_csv("D:/COS40007/Portfolio 2/final_features_EDA.csv")

print(all_data_df.head())
print("Info: ", all_data_df.info())
```

 Support Vector Machine

```python
X = all_data_df.drop(['class'], axis=1)
Y = all_data_df['class']
print("Shape of X:", X.shape)
print("Shape of y:", Y.shape)

```

10-fold cross validation mean accuracy of SVM

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
clf_73 = svm.SVC()
clf_73.fit(X_train, y_train)
Y_pred = clf_73.predict(X_test)
acc = accuracy_score(y_test, Y_pred)

print(f"Accuracy of the SVM is: {acc * 100:2f}")
```

```python

clf_cross = svm.SVC()
scores = cross_val_score(clf_cross, X, Y, cv=10)
print(scores)
```

Hyperparameter tuning

```python

param_grid = {'C': [0.1, 1, 10],
			'gamma': [1, 0.1, 0.01],
			'kernel': ['rbf']}

grid = GridSearchCV(svm.SVC(), param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(X_train, y_train)

print("Best params to fit: ", grid.best_params_)
```

```python
grid_pred = grid.predict(X_test)
print(accuracy_score(y_test, grid_pred))
```

```python
svc_with_hyp = svm.SVC(C=grid.best_params_['C'], gamma=grid.best_params_['gamma'], kernel=grid.best_params_['kernel'])
svc_with_hyp.fit(X_train, y_train)

y_pred_with_hyp = svc_with_hyp.predict(X_test)

accuracy_score_with_hyp = accuracy_score(y_test, y_pred_with_hyp)

print(f"Accuracy of the SVM with hyperparameters tuning: {accuracy_score_with_hyp * 100:2f}")
```

```python
cv_scores_with_hyp = cross_val_score(svc_with_hyp, X, Y, cv = 10)
print(f"10-fold cross validation mean accuracy score with hyperparameter tuning: {cv_scores_with_hyp.mean()*100:2f}")
     
```

```python
X_new = SelectKBest(f_classif, k=10).fit_transform(X, Y)
X_new.shape
```

```python
# 70/30 train/test split with hyperparameter tuning
X_new_train, X_new_test, y_train, y_test = train_test_split(X_new, Y, test_size=0.3, random_state=1)
grid.fit(X_new_train, y_train)
grid_pred_73 = grid.predict(X_new_test)
print(accuracy_score(y_test, grid_pred_73))
```

```python
# 10-fold class validation with hyperparameter tuning
scores_k = cross_val_score(grid, X_new, Y, cv=10)
print(scores_k)
```

```python
# Apply PCA to the original data and get the components
pca = PCA(n_components=10).fit(X)
X_pca_components = pca.transform(X)

# Display the shape of the new data after PCA
print("Shape of data after PCA (using components):", X_pca_components.shape)
```

```python
# 70/30 train/test split with hyperparameter tuning
X_pca_components_train, X_pca_components_test, y_train, y_test = train_test_split(X_pca_components, Y, test_size=0.3, random_state=1)
grid.fit(X_pca_components_train, y_train)
grid_pred_73 = grid.predict(X_pca_components_test)
print(accuracy_score(y_test, grid_pred_73))
```

```python
# 10-fold class validation with hyperparameter tuning
scores_pca = cross_val_score(grid, X_pca_components, Y, cv=10)
print(scores_pca)
```

```python
all_data_raw_df = pd.read_csv("D:/COS40007/Portfolio 2/combined_data.csv")
all_data_df = all_data_raw_df.copy()
print(all_data_df.head())
print("Info: ", all_data_df.info())
```

```python
X_data = all_data_df.drop(['class'], axis=1)
y_data = all_data_df['class']

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3, random_state=1)
```

```python
# SGDclassifier
from sklearn.linear_model import SGDClassifier
clf_sgd = SGDClassifier(max_iter=1000, tol=1e-3, random_state=1)
clf_sgd.fit(X_train, y_train)
y_pred_sgd = clf_sgd.predict(X_test)
print(accuracy_score(y_test, y_pred_sgd))
```

```python
cv_score_rf = cross_val_score(clf_sgd, X_data, y_train, cv=10)
cv_score_rf_mean = cv_score_rf.mean()
print(f"10-fold cross validation accuracy of the Random Forest model: {cv_score_rf_mean * 100:2f}")
```

```python
# RandomForest
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
clf_rf.fit(X_train, y_train)
y_pred_rf = clf_rf.predict(X_test)
print(accuracy_score(y_test, y_pred_rf))
```

```python
# MLPclassifier
from sklearn.neural_network import MLPClassifier
clf_mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
clf_mlp.fit(X_train, y_train)
y_pred_mlp = clf_mlp.predict(X_test)
print(accuracy_score(y_test, y_pred_mlp))
```
